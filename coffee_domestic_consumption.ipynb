```python
import pandas as pd
```

```python
consumption = pd.read_csv('Coffee_domestic_consumption.csv')
df_production = pd.read_csv('Coffee_production.csv')
consumption.head()
```

```python
consumption.shape
```

```python
#Let's combine the df of production and domestic consumption
combined_prod_consump_df = df_production.merge(consumption, on='Country')
selected_columns = ['Country', 'Total_production', 'Total_domestic_consumption']
result_total_prod_and_consump_df = combined_prod_consump_df[selected_columns]
```

```python
result_total_prod_and_consump_df.head()
```

```python
#What is the percentage locally consumed vs the total production
percent_consumed_over_produced = result_total_prod_and_consump_df['Total_domestic_consumption']/ result_total_prod_and_consump_df['Total_production']
percent_consumed_over_produced
```
