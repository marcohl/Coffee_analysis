```python
import matplotlib.pyplot as plt
import country_converter as coco

# Add 'Continent' column to the DataFrame
df_production['Continent'] = df_production['Country'].apply(lambda x: coco.CountryConverter().convert(names=x, to='continent'))

# Filter data for countries in America
america_data = df_production[df_production['Continent'] == 'America']

# Extract columns representing years
year_columns = america_data.columns[2:-2]

# Plotting for countries in America
plt.figure(figsize=(18, 12))
for country, row in america_data.iterrows():
    plt.plot(year_columns, row[2:-2].values, marker='o', label=str(country))

plt.title('Coffee Production Trend in America')
plt.xlabel('Year')
plt.ylabel('Production (Kg)')
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.legend(labels=america_data['Country'], loc='center left', bbox_to_anchor=(1, 0.5), title='Country', fontsize='small')
plt.grid(True)
plt.show()


```

```python
#The production plot along the years for Bolivia
# Filter data for Bolivia
bolivia_data = df_production[df_production['Country'] == 'Bolivia (Plurinational State of)']

# Extract columns representing years
year_columns = bolivia_data.columns[2:-2]

# Plotting for Bolivia
plt.figure(figsize=(12, 6))
plt.plot(year_columns, bolivia_data.iloc[0, 2:-2].values, marker='o')
plt.title('Coffee Production Trend in Bolivia')
plt.xlabel('Year')
plt.ylabel('Production (Kg)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Filter data for Bolivia, Peru, Colombia, and Ecuador
selected_countries = ['Bolivia (Plurinational State of)', 'Peru', 'Colombia', 'Ecuador']
selected_data = df_production[df_production['Country'].isin(selected_countries)]

# Sort selected data based on maximum production value
selected_data = selected_data.loc[selected_data.iloc[:, 2:-2].max(axis=1).sort_values(ascending=False).index]

# Extract columns representing years
year_columns = selected_data.columns[2:-2]

# Define colors for each country
colors = ['blue', 'green', 'red', 'purple']

# Plotting for selected countries
plt.figure(figsize=(12, 6))
for idx, (country, row) in enumerate(selected_data.iterrows()):
    plt.plot(year_columns, row[2:-2].values, marker='o', label=country, color=colors[idx])

# Create a custom legend for selected countries
legend_labels = selected_data['Country']
plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), title='Country', fontsize='small', labels=legend_labels)

plt.title('Coffee Production Trend (Bolivia, Peru, Colombia, Ecuador)')
plt.xlabel('Year')
plt.ylabel('Production (Kg)')
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.grid(True)
plt.tight_layout()  # Ensures the plot layout is tight
plt.show()
```

```python
# Assuming df_production is your original DataFrame

# Combine 'Arabica/Robusta' and 'Robusta/Arabica' into a single category 'Arabica/Robusta'
df_production['Coffee type'] = df_production['Coffee type'].replace({'Robusta/Arabica': 'Arabica/Robusta'})

# Calculate total production for each 'Coffee type'
total_production_by_type = df_production.groupby('Coffee type')['Total_production'].sum()

# Calculate the percentage of total production for each 'Coffee type'
percentage_by_type = (total_production_by_type / total_production_by_type.sum()) * 100

# Display the result
print(percentage_by_type)
```

```python
## How was the coffee production last year available? (2019/20)

# Select the last year column
last_year_column = df_production.columns[-2]

# Sort the DataFrame by total production in the last year
sorted_df = df_production.sort_values(by=last_year_column, ascending=False)

# Plotting
plt.figure(figsize=(14, 8))
plt.bar(sorted_df['Country'], sorted_df[last_year_column], color='skyblue', edgecolor='black')

plt.title(f'Coffee Production by Country (Stacked) in {last_year_column}')
plt.xlabel('Country')
plt.ylabel('Total Production (Kg)')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Ensures the plot layout is tight
plt.show()


```

```python
#who were the top 5 coffee producers for each year?
# Extract columns representing years
year_columns = df_production.columns[2:-1]

# List to store top 5 producers for each year
top_producers_by_year = []

# Loop through each year
for year in year_columns:
    # Sort DataFrame by production for the current year in descending order
    sorted_df = df_production.sort_values(by=year, ascending=False)
    
    # Extract the top 5 producers for the current year
    top_5_producers = sorted_df[['Country', year]].head(5).values.tolist()
    
    # Append the result to the list
    top_producers_by_year.append((year, top_5_producers))

# Display the result
for year, top_producers in top_producers_by_year:
    print(f"Top 5 Producers in {year}:\n")
    for producer in top_producers:
        print(f"{producer[0]}: {producer[1]:,.0f} Kg")
    print("\n")

```
